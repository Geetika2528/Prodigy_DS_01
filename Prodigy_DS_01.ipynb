```python
import os
import pandas as pd
import matplotlib.pyplot as plt
directory = r"C:\Users\ASUS\Prodigy_InfoTech\POPULATION CHART"
files = os.listdir(directory)
csv_files = [f for f in files if f.endswith('.csv')]
if not csv_files:
    raise FileNotFoundError("No CSV files found in the directory")
file_path = os.path.join(directory, csv_files[0])
df = pd.read_csv(file_path, skiprows=4)
print(df.head())
df_filtered = df[['Country Name', '2022']].rename(columns={'Country Name': 'Country', '2022': 'Population_2022'})
df_filtered.dropna(subset=['Population_2022'], inplace=True)
df_filtered['Population_2022'] = pd.to_numeric(df_filtered['Population_2022'], errors='coerce')
print(df_filtered.head())
plt.figure(figsize=(12, 8))
plt.hist(df_filtered['Population_2022'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Population in 2022')
plt.ylabel('Number of Countries')
plt.title('Population Distribution of Countries in 2022')
plt.show()
```
